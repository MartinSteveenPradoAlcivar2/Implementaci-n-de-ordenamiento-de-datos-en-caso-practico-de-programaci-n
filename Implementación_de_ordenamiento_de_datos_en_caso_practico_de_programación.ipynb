```python
#Martin Prado

#Parte 1

#Es la base principal desde la cual se muestra el menú
#Todo por medio del nombre del plato como clave y su información
menu_restaurant = {
    'Breaded Shrimp': {
        'Price': 7.00,
        'Ingredients': ('Rice', 'Shrimp', 'Beans', 'Salad')
    }, 'Ceviche Shrimp': {
        'Price': 7.50,
        'Ingredients': ('Oil', 'Shrimp', 'Onion', 'Water')
    }, 'Breaded Fish': {
        'Price': 8.00,
        'Ingredients': ('Fish', 'Rice', 'Beans', 'Oil')
    }, 'Chicken Frice': {
        'Price': 5.00,
        'Ingredients': ('Rice', 'Chicken', 'Oil', 'Beans')
    }, 'Encocao Fish': {
        'Price': 8.50,
        'Ingredients': ('Rice', 'Coconut', 'Fish', 'Ripe Banana')
    }
}

#Muestra un menú con los platos y su información respectiva
#En el orden establecido
print(' Menu ')
for plato, platoI in menu_restaurant.items():
    print(f"{plato}")
    print(f"  Price: {platoI['Price']}")
    print(f"  Ingredients: {(platoI['Ingredients'])}")
    print('-'*50)


#Parte 2
#Esta parte del código es la que interactúa con el usuario
#Guardando el pedido y dando un resumen y un total
print('Por favor ingrese el plato que guste, o escriba "salir" para finalizar: ')
plato2 = []
precioF = 0.0

while True:
#Aquí empieza el bucle que se ejecutará hasta que el usuario ingrese “salir”
#Y por medio del “break” el bucle se rompa
  pedido = input('Introduzca el plato deseado: ')
  if pedido.lower() == 'salir':
    print('Pedido Finalizado')
    break
    print('-'*50)

  if pedido in menu_restaurant:
#Los platos que ingrese el usuario serán guardados junto a su información
#Siempre y cuando los platos estén en el menú
    plato2.append(pedido)
    precioF += menu_restaurant[pedido]['Price']
    print(f"'{pedido}' a sido añadido al pedido.")

#Si el plato ingresado no está en el menú será un dato invalido
  else: print('Ese plato no esta en el menú, lo sentimos.')

#Se imprime la lista con los platos ingresados
#Junto al valor total sumado de todos los platos
print('Resumen de su pedido: ')
if plato2:
  for platoR in plato2:
      print(f'{platoR}')
  print(f'El total es de ${precioF} dólares.')
else: print('No hay un pedido existente.')
print('-'*50)

#Parte 3
#En este apartado se modifica el precio de un plato de manera directa
#Por medio de la clave que es su nombre en la lista principal
platoCam = menu_restaurant ['Breaded Shrimp']
menu_restaurant['Breaded Shrimp']['Price'] = 8.50
for plato, platoI in menu_restaurant.items():
    print(f"{plato}")
    print(f"  Price: {platoI['Price']}")
    print(f"  Ingredients: {(platoI['Ingredients'])}")
    print('-'*50)

#En este apartado se añade directamente un plato clave junto a su información
menu_restaurant['El plato por tiempo limitado es: Bacon Burger'] = {
        'Price': 6.50,
        'Ingredients': ('Bacon','Lettuce', 'Bread', 'Meat')
}
for plato, platoI in menu_restaurant.items():
    print(f"{plato}")
    print(f"  Price: {platoI['Price']}")
    print(f"  Ingredients: {(platoI['Ingredients'])}")
    print('-'*50)

#Aquí se crea una lista que se muestra separa
#En la cual está el menú del día
#Mostrando los platos escogidos directamente del menú principal
print('Especialidad de la casa: Platos del día: ')
platoD = ['Ceviche Shrimp', 'Chicken Frice']
for platoDia in platoD:
  if platoDia in menu_restaurant:
    platoInf = menu_restaurant[platoDia]
    print(f'{platoDia}')
    print(f" Price: {platoInf['Price']}")
    print(f" Ingredients: {platoInf['Ingredients']}")
    print('-'*50)

#Parte 4
#En este apartado se muestran los 3 errores de estructura de datos
menu_restaurante = {
    'Breaded Shrimp': {
        'Price': 7.00,
        'Ingredients': ('Rice', 'Shrimp', 'Beans', 'Salad')
    },
    'Ceviche Shrimp': {
        'Price': 7.50,
        'Ingredients': ('Oil', 'Shrimp', 'Onion', 'Water')
    },
    'Breaded Fish': {
        'Price': 8.00,
        'Ingredients': ('Fish', 'Rice', 'Beans', 'Oil')
    },
    'Chicken Frice': {
        'Price': 5.00,
        'Ingredients': ('Rice', 'Chicken', 'Oil', 'Beans')
    },
    'Encocao Fish': {
        'Price': 8.50,
        'Ingredients': ('Rice', 'Coconut', 'Fish', 'Ripe Banana')
    }
}
#IndexError
#Es un error de tupla que aparece al momento de buscar un elemento que no existe en la lista
#indexE = menu_restaurante['Breaded Shrimp']['Ingredients'][5]

#KeyError
#Se da al momento de buscar una clave que no existe en la lista
#KeyE = menu_restaurante['Camaron Apanado']

#TypeError
#Se da cuando se trata de ejecutar una orden que no es vista como valida
#En algún tipo de dato
#menu_restaurante['Ceviche Shrimp']['Ingredients'][1] = 'Water'
```
